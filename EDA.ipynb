```python
df['Age'].fillna((df['Age'].mean()), inplace=True)
#fill Nan values with  age is a numerical values so we can use  (mean) or (median)
```

```python
df.isnull().sum()
# find how many are null values
```

```python
sns.set_style("whitegrid")
sns.boxplot(x="Pclass",y="Age",data = df)
#to visualize age and pclass using boxplot
```

```python
df['is_Cabin'] = df['Cabin'].apply(lambda x:0 if pd.isnull(x) else 1)
#to fill Nanvalues 0 else 1.  lambda is a conties values like 1 2 3 4 5
```

```python
df.head(2)
#show values of 2
```

```python
df = df.drop('Cabin', axis=1)
# removes cabin column rows=0 and columns=1
```

```python
df.isnull().sum()
# checking null values how many are still remaining
```

```python
df['Embarked'].fillna((df['Embarked'].mode()[0]), inplace=True)
#fill Nan values with  Embarked is a categorical values so we can use  mode  [0] repeat v
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df = df.drop(['PassengerId','Name','Ticket'], axis=1)
#remove all the columns from the data set using drop
```

```python
df.head(2)
```

```python
df['Sex']=df['Sex'].apply(lambda X:1 if X=="male" else 0)
#sex mean male or female it will changeed 1 or 0 using lambda function
```

```python
df.head(3)
```

```python
df['Embarked'].unique()
#checking how many unique values in Embarked
```

```python
df['Embarked'].nunique()
```

```python
df['Embarked'].value_counts()
```

```python
v =pd.get_dummies(df['Embarked'])
#checking dummies in embarked column and stored into v
```

```python
v
```

```python
df = pd.concat([df,v],axis=1)
df.head(2)
#concentrate in v into df in column
```

```python
df.drop(['Embarked'],axis=1,inplace = True)
#removed  Embarked column from the dataset
```
